## Задание 2
По данным файла `power.csv` посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
quant = pd.read_csv('power.csv')
```

```python
quant.head()
```

```python
countries = ['Latvia', 'Estonia', 'Lithuania']
categories = [4, 12, 21]
```

```python
pribalt = quant[quant.country.isin(countries) &
          quant.category.isin(categories) &
          (2005 <= quant.year) &
          (quant.year <= 2010) &
          (quant.quantity >= 0)]
pribalt.quantity.sum()
```

```python
pribalt.head()
```

```python
pribalt.pivot_table(index='country', columns='year', values='quantity', aggfunc='sum', fill_value=0, margins=True)
```
